Task 6: Rename columns wherever necessary so that all datasets have a common column name for merging.

```python
gdp.rename(columns={"Country":"country"},inplace=True)
gdp
```

Task 6: Rename columns wherever necessary so that all datasets have a common column name for merging.

```python
literacy.rename(columns={"name":"country"},inplace=True)
literacy
```

Task 8: Merge all four datasets into one DataFrame.

```python
mergedd=pd.merge(population,gdp,on="country",how="inner")
```

```python
mergedd=pd.merge(mergedd,user,on="country",how="inner")
```

```python
mergedd=pd.merge(mergedd,literacy,on="country",how="inner")
mergedd
```

Task 9: Create the following calculated columns:
1. Internet Penetration (%)
2. GDP Per Capita

```python
mergedd["Internet_Penetration"]=(mergedd["internet_users"]/mergedd["population"]*100).round(2)
mergedd


```

```python
mergedd["GDP_Per_Capita"]=(mergedd["GDP"]/mergedd["population"]).round(2)
mergedd


```

Task 10: Create a SQLite database and store the final merged dataset into a table named 'country_statistics'.

```python
import sqlite3
import pandas as pd

```

```python
conn = sqlite3.connect("country_statistics.db")
```

```python
mergedd.to_sql(
    "country_statistics",
    conn,
    if_exists="replace",
    index=False
)
```
